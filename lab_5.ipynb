```python
import pandas
import matplotlib
import scipy
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("car_crashes.csv")
# print(data.head())
D=data.head()
print(D)
```

```python
#SCATTER PLOT
plt.scatter(D.alcohol,D.ins_losses)
plt.show()
```

```python
df=sns.load_dataset('car_crashes')
x = df.speeding
y = df.alcohol
plt.xlabel('spedding')
plt.ylabel('alcohol')
plt.title('Bar Graph Example')
plt.bar(x,y)
plt.show()
```

```python

sns.boxplot(x='ins_premium',y='ins_losses',data=df)
plt.show()
```

```python
plt.hist(data['alcohol'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('alcohol')
plt.ylabel('speeding')
plt.title('Histogram Example')
```
